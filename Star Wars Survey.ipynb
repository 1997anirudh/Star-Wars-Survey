# Star Wars Survey

Survey conducted to answer 1 question, this survey recieved 835 responses.. 
The data is stored in an excel file given [here](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey)

Some of the columns in the dataset
- RespondentID - An anonymized ID for the respondent (person taking the survey)
- Gender - The respondent's gender
- Age - The respondent's age
- Household Income - The respondent's income
- Education - The respondent's education level
- Location (Census Region) - The respondent's location
- Have you seen any of the 6 films in the Star Wars franchise? - Has a Yes or No response
- Do you consider yourself to be a fan of the Star Wars film franchise? - Has a Yes or No response



```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv",encoding="ISO-8859-1")
star_wars.head()
star_wars.columns
```

```python
star_wars = star_wars[star_wars["RespondentID"].notnull()]
star_wars
```

```python
print(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts(dropna = False))
print(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts(dropna = False))
```

```python
def yes_no_map(string):
    if(string == "Yes"):
        return True
    elif (pd.isnull(string)):
        return np.nan
    else:
        return False
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].apply(yes_no_map)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].apply(yes_no_map)
star_wars[["Have you seen any of the 6 films in the Star Wars franchise?","Do you consider yourself to be a fan of the Star Wars film franchise?"]]
```

```python
col_dict = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': "seen_ep1",
    'Unnamed: 4': "seen_ep2",
    "Unnamed: 5": "seen_ep3",
    "Unnamed: 6": "seen_ep4",
    "Unnamed: 7": "seen_ep5",
    "Unnamed: 8": "seen_ep6"
}

star_wars = star_wars.rename(columns=col_dict)
```

```python
def film_dict(movie_name):
    if(movie_name == 'Star Wars: Episode I  The Phantom Menace'):
        return True
    elif(movie_name == 'Star Wars: Episode I  The Phantom Menace'):
        return True
    elif(movie_name =='Star Wars: Episode II  Attack of the Clones'):
        return True
    elif(movie_name =='Star Wars: Episode III  Revenge of the Sith'):
        return True
    elif(movie_name == 'Star Wars: Episode IV  A New Hope'):
        return True
    elif(movie_name =='Star Wars: Episode V The Empire Strikes Back'):
        return True
    elif(movie_name =='Star Wars: Episode VI Return of the Jedi'):
        return True
    else:
        return False
```

```python
star_wars[star_wars.columns[3:9]] = star_wars[star_wars.columns[3:9]].applymap(film_dict)
star_wars[star_wars.columns[3:9]]
```

```python
#renaming columns
renaming_15_29 = {
    "Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.": "Han Solo",
    "Unnamed: 16": "Luke Skywalker",
    "Unnamed: 17": "Princess Leia Organa",
    "Unnamed: 18": "Anakin Skywalker",
    "Unnamed: 19": "Obi Wan Kenobi",
    "Unnamed: 20": "Emperor Palpatine",
    "Unnamed: 21": "Darth Vader",
    "Unnamed: 22": "Lando Calrissian",
    "Unnamed: 23": "Boba Fett",
    "Unnamed: 24": "C-3P0",
    "Unnamed: 25": "R2 D2",
    "Unnamed: 26": "Jar Jar Binks",
    "Unnamed: 27": "Padme Amidala",
    "Unnamed: 28": "Yoda"
}
star_wars = star_wars.rename(columns=renaming_15_29)
```

To find the most loved and hated characters unanimously, let us consider *Very favorably* and *Very unfavorably* as the others aren't sure shot answers. 

```python
liked_list = { }
disliked_list = { }
controversial_list = { }
for i in range(15,29):
    liked = star_wars[star_wars[star_wars.columns[i]] == "Very favorably"]
    disliked = star_wars[star_wars[star_wars.columns[i]] == "Very unfavorably"]
    liked_size = liked.size
    disliked_size = disliked.size
    controversial = liked_size/disliked_size
    liked_list[star_wars.columns[i]] = liked_size
    disliked_list[star_wars.columns[i]] = disliked_size
    controversial_list[star_wars.columns[i]] = controversial 
#     liked_list.append(liked_size)
#     disliked_list.append(disliked_size)
#     controversial_list.append(controversial)
print(liked_list)
print("\n")
print(disliked_list)
print("\n")
print(controversial_list)
```

```python
fav_unfav = pd.DataFrame.from_dict([liked_list,disliked_list,controversial_list]) 
fav_unfav
```

A deeper analysis using education and gender ought to answer a lot more questions, regarding the characters.

```python
male_character = star_wars[star_wars["Gender"] == "Male"][star_wars.columns[15:29]]
female_character = star_wars[star_wars["Gender"] == "Female"][star_wars.columns[15:29]]
print("Male respondents")
print(male_character.shape[0])
print("Female respondents")
print(female_character.shape[0])
def gender_character(inp):
    if(inp == "Very favorably"):
        return 1
    elif(inp == "Very unfavorably"):
        return -1
male_character = male_character.applymap(gender_character)
female_character = female_character.applymap(gender_character)
male_sum = male_character.sum()
female_sum = female_character.sum()
# print(male_sum)
# print(female_sum)
gender_df = pd.DataFrame()
gender_df["Character liked by Men"] = male_sum
gender_df["Character liked by Women"] = female_sum
gender_df   
```

There are more women than men in this survey, 
characters like Yoda, Solo, Skywalker and Leia are loved most by both genders. 
Saem with R2-D2 and C-3P0. 

Men tend to like Vader a little bit more than the woman. 
But Jar Jar Binks is disliked by the men a lot more in comparison to the women.
